```python
def GetFileNames():
    #Getting all file names and storing them in a list
    path = os.getcwd()
    path = path+"\\Comments"
    AllFilePaths=[]
    for item in os.listdir(path):
        if not os.path.isfile(os.path.join(path, item)):
            pass
        else:
            AllFilePaths.append(item)
    print (color.BOLD+"Number of Comment Text-files"+color.RED+":"+color.END,len(AllFilePaths))
    return AllFilePaths


def RemoveDuplicates(l):
    s=[]
    for i in l:
        if i not in s:
            s.append(i)
    return s   
#Defining some Colors for Text
class color:
    HEADER = '\033[120m'
    OKBLUE = '\033[94m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'

#Defining Some Basic Outlines For Printing
def ColorText(headings='',data='',data2=''):
    #Converts Type
    try:
        if(type(headings)!=str):
            H=str(headings)
        if(type(data)!=str):
            D=str(data)
    except:
        Message=color.GREEN+"ColorText: Unable to convert input datatype to string!"+color.END
        print(color.RED+color.BOLD+color.UNDERLINE+"ERROR : "+color.END+Message)
#Prints only Header
    H=color.BOLD+color.UNDERLINE+color.RED+headings+color.END
    D=color.BLUE+data+color.END
    D1=color.GREEN+data2+color.END
    return H,D,D1
```

```python
FileNames=GetFileNames()
'''
Variables         |   Function
------------------------------------
    Output        |   The Complete Brand-Mobile-Comments Dictionary    
    Brands        |   All names of Brands
    Model_ID      |   Model ID for a certain Brand Brand
    Data          |   File Directory List of All-Comments
    Labels        |   Brand, Model,
Objectives:
      1)Taking out the text file format from string
      2)Extracting Brands and Model Names from Text
      3)Splitting, Replacing, Extracting Space
Returns: 
          Count-Dictionary for Number Of Comments:
          Indexing Method: Output[Brand][Model][Count_Of_Comments]
'''
def GetBMC(Data):
    Output={}
    Model_ID=0
    Brands=[]
    SelectBrand=""
    for Raw in Data:
        Text=Raw.replace(".txt","")
        Brand=Text.split(' ', 1)[0]
        if(Brand[:1].islower()):
            Brand=Brand[0:1].upper()+Brand[1:]
        Model=Text.replace(Brand,"")
        if(Model[:1]==' '):
            Model=Model[1:]
        elif(Model[:1].lower()):
            Model=Model[:1].upper()+Model[1:]
        if(Brand!=SelectBrand):
            SelectBrand=Brand
            Brands.append(Brand)
        if (Brand not in Output):
            Model_ID=0
            Output.setdefault(Brand,{})
        else:
            Output[Brand].setdefault(Model_ID,Model)
            Model_ID+=1
    Brand_ID=0
    for Brand in Brands:
        Output.setdefault(Brand,{})
        Brand_ID+=1
    return Output,Brands

'''
Objectives:
    BrandModelCounts prints Number of Models per Brand
'''
def BrandModelCounts(BMC,Brands):
    Heading=" Brand  \t | \t Model-Counts "
    Head,Brands,Data=ColorText(Heading,'','')
    Brands=Data=''
    print (color.HEADER+Heading+color.END)
    for Brands in BMC:
        t=Brands
        x=len(BMC[Brands])
        x=str(x)
        S,Brands,Data=ColorText('',Brands,x)
        if(len(t)<5):
            print (" ",Brands ,"\t \t \t",Data)
        else:
            print (" ",Brands ,"\t \t",Data)

'''
Objective: 
      To Print:
      1)No of Comments Per Brand
      2)No of Comments Per Phone from a Brand
    
'''
def BMCommentCount(BMC):
    for Brands in BMC:
        for Phones in Brands:
            for Comments in Phones:
                print (color.BLUE+str(Comments)+color.END)
        
```

```python
import pickle

def np_SaveData(myList,filename):
    np.array(myList).dump(open(filename+'.npy', 'wb'))
    
def np_LoadData(filename):
    myArray = np.load(open(filename+'.npy', 'rb'))
    return myArray

def SaveDictionary(Dic,DicName):
    #This saves all the key-value pair dictionaries
    pickle_out = open(DicName,"wb")
    pickle.dump(Dic, pickle_out)
    pickle_out.close()
    
def ReadDictionary(DicName):
    pickle_in = open(DicName,"rb")
    diction = pickle.load(pickle_in)
    return diction
```

1) Seed Extraction 

    Dataset of User Comments is used to extract sentimental seeds

2) Seed Expansion

    A synomnymy graph is build based on synonyms, Then, Graph propagation algorithm is used to get general sentimal words

3) DSSW Extraction

    Two kinds of patterns are generated automatically, which are furthur used to extract new sentimental words from corpus


```python
def TagWords(Corpus):
    CorpusTagged=[]
    Files=len(Corpus)
    counter=0
    for Phones in Corpus:
        Tags=[]
        for Comments in Phones:
            text=word_tokenize(Comments)
            x=nltk.pos_tag(text)
            Tags.append(x)
        CorpusTagged.append(Tags)
        print("Progress {:2.1%}".format(counter / Files), end="\r")
        counter+=1
    return CorpusTagged
```

```python
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
n_instances=100
subj_docs=[(sent,'subj') for sent in subjectivity.sents(categories='subj')[:n_instances]]
obj_docs=[(sent,'obj') for sent in subjectivity.sents(categories='obj')[:n_instances]]
print (len(subj_docs),len(obj_docs))
train_subj_docs = subj_docs[:80]
test_subj_docs = subj_docs[80:100]
train_obj_docs = obj_docs[:80]
test_obj_docs = obj_docs[80:100]
training_docs = train_subj_docs+train_obj_docs
testing_docs = test_subj_docs+test_obj_docs
```

```python
sentim_analyzer = SentimentAnalyzer()
all_words_neg = sentim_analyzer.all_words([mark_negation(doc) for doc in training_docs])

unigram_feats = sentim_analyzer.unigram_word_feats(all_words_neg, min_freq=4)
len(unigram_feats)
sentim_analyzer.add_feat_extractor(extract_unigram_feats, unigrams=unigram_feats)

training_set = sentim_analyzer.apply_features(training_docs)
test_set = sentim_analyzer.apply_features(testing_docs)

trainer = NaiveBayesClassifier.train
classifier = sentim_analyzer.train(trainer, training_set)

for key,value in sorted(sentim_analyzer.evaluate(test_set).items()):
    print('{0}: {1}'.format(key, value))
```
